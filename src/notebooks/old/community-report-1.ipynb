```python
import os # creating directories
import pandas as pd
import matplotlib.pyplot as plt
import csv # remove str quotation marks
```

```python
# load data
df = pd.read_csv("../datain/community/copy.net")
h = df.head()
h
```

```python
df = pd.DataFrame(list(df["from    to  mentions"].str.split("\t")), columns=["from", "to", "mentions"])
df
```

```python
# convert type to int
df["from"] = df["from"].astype(int)
df.to = df.to.astype(int)
df.mentions = df.mentions.astype(int)
```

```python
# format weight column in format that networkx takes as input
df['weight'] = "{'weight':" + df['mentions'].astype(str) + "}"
df
```

```python
# ouput to csv
selected_columns = ["from", "to", "weight"]
df.to_csv("../datain/community/edges.csv", index=False, sep=" ", quoting=csv.QUOTE_NONE, quotechar="", escapechar="\\", columns = selected_columns)
```
