# Sentiment Analysis on Specific Topics

Currently set to analyse topic 4's sentiment (see graphs below for results)

```python
# dependencies - install first time using this notebook
# !pip3 install vaderSentiment
# !pip3 install --upgrade pip
# !pip3 install xlrd
# !pip3 install openpyxl

import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```python
df_topic = pd.read_excel('datain/sentiment/grouped-by-topic_Topics0-10_Sheets.xlsx', sheet_name='Topic4')
df_topic
```

```python
analyzer = SentimentIntensityAnalyzer()
for tweet in df_topic['Tweet']:
    vs = analyzer.polarity_scores(tweet)
    print("{:-<65} {}".format(tweet, str(vs)))
```

```python
df_topic['Tweet'].index
```

```python
#https://github.com/sidneykung/twitter_hate_speech_detection/blob/master/preprocessing/VADER_sentiment.ipynb
pol = lambda x: analyzer.polarity_scores(x)
df_topic['Polarity'] = df_topic['Tweet'].apply(pol)
df_topic
```

```python
df = pd.concat([df_topic.drop(['Polarity'], axis=1), df_topic['Polarity'].apply(pd.Series)], axis=1)
df
```

```python
plt.plot(df.index, df['compound'].sort_values())
plt.title('Sentiment Analysis using Vader')
plt.ylabel('Polarity')
plt.xlabel('Tweets')
```

```python
def calc_polarity(x, bound):
    if x < -0.05:
        return -1
    elif x > 0.05:
        return 1
    else:
        return 0

round_pol = lambda x: calc_polarity(x, 0.05)
```

```python
df['compound'][0]
```

```python
df['Polarity'] = df['compound'].apply(round_pol)
df
```

```python
plt.plot(df.index, df['Polarity'].sort_values())
plt.title('Sentiment Analysis using Vader')
plt.ylabel('Polarity')
plt.xlabel('Tweets')
```

```python
sentiments = df.groupby('Polarity').count()
sentiments
```

```python
plt.bar(sentiments.index, sentiments["compound"])
plt.title('Sentiment')
plt.xlabel('Polarity')
plt.ylabel('Count')
```
